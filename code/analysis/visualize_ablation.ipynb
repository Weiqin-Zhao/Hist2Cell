```python
import matplotlib.pyplot as plt
from PIL import Image
Image.MAX_IMAGE_PIXELS = None


# 打开四张jpg图片
img_A = Image.open('/data2/r10user3/Spatial_Gene_Cell_Ratio/data/human_lung_cell2location/WSA_LngSP8759311/WSA_LngSP8759311.jpg')
img_B = Image.open('/data2/r10user3/Spatial_Gene_Cell_Ratio/data/human_lung_cell2location/WSA_LngSP8759311/cell_abundance_figure/q05cell_abundance_w_sf_Ciliated_ground_truth.jpg')
img_C = Image.open('/data2/r10user3/Spatial_Gene_Cell_Ratio/data/human_lung_cell2location/WSA_LngSP8759311/cell_abundance_figure/q05cell_abundance_w_sf_Ciliated_single_task.jpg')
img_D = Image.open('/data2/r10user3/Spatial_Gene_Cell_Ratio/data/human_lung_cell2location/WSA_LngSP8759311/cell_abundance_figure/q05cell_abundance_w_sf_Ciliated_multi_task.jpg')

img_A = img_A.resize((img_A.width//20, img_A.height//20))
img_B = img_B.resize((img_B.width//20, img_B.height//20))
img_C = img_C.resize((img_C.width//20, img_C.height//20))
img_D = img_D.resize((img_D.width//20, img_D.height//20))
```

```python
# 设置Matplotlib的风格为ggplot
plt.style.use('ggplot')

# 设置字体
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 14

# 创建一个1x3的网格布局，并调整子图之间的间距和边距
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 7))
fig.subplots_adjust(wspace=0.02, hspace=0.05, left=0.05, right=0.95, top=0.8, bottom=0.2)

# 在网格的第一个位置显示图片1，并设置标题为"图A"
axes[0].imshow(img_A)
axes[0].set_title('Whole Slide Image', fontsize=15)
axes[0].axis('off')  # 隐藏坐标轴

# 在网格的第二个位置显示图片2，并设置标题为"图B"
axes[1].imshow(img_B)
axes[1].set_title('Ground Truth', fontsize=15)
axes[1].axis('off')  # 隐藏坐标轴

# 在网格的第三个位置显示图片3，并设置标题为"图C"
axes[2].imshow(img_C)
axes[2].set_title('Single-task GNN', fontsize=15)
axes[2].axis('off')  # 隐藏坐标轴

axes[3].imshow(img_D)
axes[3].set_title('Multi-task GNN', fontsize=15)
axes[3].axis('off')  # 隐藏坐标轴

# 添加整个图像的标题
fig.suptitle('Cell Type Ciliated Abundance', fontsize=15, y=0.9)

# 显示整个网格布局
plt.show()
```
