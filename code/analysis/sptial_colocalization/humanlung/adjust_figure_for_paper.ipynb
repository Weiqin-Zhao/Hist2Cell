```python
from tkinter import font
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

case = "A37"
slide = "WSA_LngSP10193347"

df_label = pd.read_csv(os.path.join("/data1/r20user3/shared_project/Hist2Cell/code/analysis/sptial_colocalization/humanlung", case, slide, "MoranR_Cell2location.csv"), index_col=0)
df_figure = df_label[['A', 'B', 'R_val']]
correlation_df = df_figure.pivot(index="A", columns="B", values="R_val")
for var in correlation_df.index:
    correlation_df.at[var, var] = 1
    correlation_df = correlation_df.combine_first(correlation_df.T)
correlation_df['gdT'].gdT = 1.0
cell2location_cluster = sns.clustermap(correlation_df, figsize=(30, 30), cmap="viridis")
cax = cell2location_cluster.ax_cbar
cax.tick_params(labelsize=35)
row_linkage = cell2location_cluster.dendrogram_row.linkage
col_linkage = cell2location_cluster.dendrogram_col.linkage
# 获取热图轴并调整x轴和y轴字体大小
heatmap_ax = cell2location_cluster.ax_heatmap
heatmap_ax.tick_params(axis='x', labelsize=18)  # 调整x轴字体大小
heatmap_ax.tick_params(axis='y', labelsize=18)  # 调整y轴字体大小
plt.savefig(os.path.join("/data1/r20user3/shared_project/Hist2Cell/code/analysis/sptial_colocalization/humanlung/selected_figures", case+"_"+slide+"_"+"cell2location_cluster_viridis.pdf"), bbox_inches='tight', pad_inches=0)
plt.show()
plt.close()

df_label = pd.read_csv(os.path.join("/data1/r20user3/shared_project/Hist2Cell/code/analysis/sptial_colocalization/humanlung", case, slide, "MoranR_Hist2Cell.csv"), index_col=0)
df_figure = df_label[['A', 'B', 'R_val']]
correlation_df = df_figure.pivot(index="A", columns="B", values="R_val")
for var in correlation_df.index:
    correlation_df.at[var, var] = 1
    correlation_df = correlation_df.combine_first(correlation_df.T)
correlation_df['gdT'].gdT = 1.0
hist2cell_cluster = sns.clustermap(correlation_df, figsize=(30, 30), row_linkage=row_linkage, col_linkage=col_linkage, cmap="viridis")
cax = hist2cell_cluster.ax_cbar
cax.tick_params(labelsize=35)
# 获取热图轴并调整x轴和y轴字体大小
heatmap_ax = hist2cell_cluster.ax_heatmap
heatmap_ax.tick_params(axis='x', labelsize=18)  # 调整x轴字体大小
heatmap_ax.tick_params(axis='y', labelsize=18)  # 调整y轴字体大小
plt.savefig(os.path.join("/data1/r20user3/shared_project/Hist2Cell/code/analysis/sptial_colocalization/humanlung/selected_figures", case+"_"+slide+"_"+"hist2cell_cluster_viridis.pdf"), bbox_inches='tight', pad_inches=0)
plt.show()
plt.close()

df_label = pd.read_csv(os.path.join("/data1/r20user3/shared_project/Hist2Cell/code/analysis/sptial_colocalization/humanlung", case, slide, "MoranR_STNet.csv"), index_col=0)
df_figure = df_label[['A', 'B', 'R_val']]
correlation_df = df_figure.pivot(index="A", columns="B", values="R_val")
for var in correlation_df.index:
    correlation_df.at[var, var] = 1
    correlation_df = correlation_df.combine_first(correlation_df.T)
correlation_df['gdT'].gdT = 1.0
stnet_cluster = sns.clustermap(correlation_df, figsize=(30, 30), row_linkage=row_linkage, col_linkage=col_linkage, cmap="viridis")
cax = stnet_cluster.ax_cbar
cax.tick_params(labelsize=35)
# 获取热图轴并调整x轴和y轴字体大小
heatmap_ax = stnet_cluster.ax_heatmap
heatmap_ax.tick_params(axis='x', labelsize=18)  # 调整x轴字体大小
heatmap_ax.tick_params(axis='y', labelsize=18)  # 调整y轴字体大小
plt.savefig(os.path.join("/data1/r20user3/shared_project/Hist2Cell/code/analysis/sptial_colocalization/humanlung/selected_figures", case+"_"+slide+"_"+"stnet_cluster_viridis.pdf"), bbox_inches='tight', pad_inches=0)
plt.show()
plt.close()
```

```python
# load the 3 saved png files and combine them into one figure

import os
from PIL import Image
import matplotlib.pyplot as plt

image_files = ["/data1/r20user3/shared_project/Hist2Cell/code/analysis/sptial_colocalization/humanlung/selected_figures/A37_WSA_LngSP10193347_cell2location_cluster.png", 
               "/data1/r20user3/shared_project/Hist2Cell/code/analysis/sptial_colocalization/humanlung/selected_figures/A37_WSA_LngSP10193347_hist2cell_cluster.png", 
               "/data1/r20user3/shared_project/Hist2Cell/code/analysis/sptial_colocalization/humanlung/selected_figures/A37_WSA_LngSP10193347_stnet_cluster.png"]
images = []
widths = []
heights = []
titles = ["Cell2location", "Hist2Cell", "STNet"]

# Load images and get dimensions
for image_file in image_files:
    image = Image.open(image_file)
    width, height = image.size
    widths.append(width)
    heights.append(height)
    images.append(image)

# Calculate the total width and maximum height
total_width = sum(widths)
max_height = max(heights)

# Create a figure with the total width and maximum height
fig = plt.figure(figsize=(total_width / 80, max_height / 80), dpi=80)

# Add images to the figure
for idx, image in enumerate(images):
    ax = fig.add_axes([sum(widths[:idx]) / total_width, 0, widths[idx] / total_width, heights[idx] / max_height])
    ax.imshow(image)
    ax.axis("off")
    # ax.set_title(titles[idx], fontsize=80, y=1.02)

plt.show()
```
