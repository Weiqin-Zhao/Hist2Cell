```python
import glob
import pandas as pd
import os

samples_path = [path for path in glob.glob("/data1/r20user3/shared_project/Hist2Cell/data/human_lung_cell2location/*") if not path.endswith(".h5ad") and not path.endswith(".xlsx")]
df_list = list()

for sample in samples_path:
    df_list.append(pd.read_csv(os.path.join(sample, "cell_ratio.csv"), index_col=0))
    
df_merge = pd.concat(df_list, axis=0)
df_merge
```

```python
abundace_sum = df_merge.sum(axis=0)
pd.DataFrame(abundace_sum, columns=["abundance"]).sort_values(by="abundance", ascending=False).index.tolist()
```
