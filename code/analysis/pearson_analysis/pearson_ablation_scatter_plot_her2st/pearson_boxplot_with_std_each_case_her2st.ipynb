```python
from glob import glob
tif_list = glob('/data1/r20user3/shared_project/Hist2Cell/code/training/train_test_splits/her2st/test*')
tif_list.sort()
test_slides = list()
for tif in tif_list:
    tif_path = tif.split('_')[-1].split('.')[0]
    test_slides.append(tif_path)
test_slides
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.graph_objs as go

cell_names = [cell[23:] for cell in list(pd.read_csv("/data1/r20user3/shared_project/Hist2Cell/data/her2st/A1/cell_ratio.csv").columns)[1:]]
for case in test_slides:
    with open('/data1/r20user3/shared_project/Hist2Cell/code/training/pearson_log/her2st_epoch100_lr1e-4_2hop_ensemble_onlycell_'+case+'_cell_pearson_list.pkl', 'rb') as f:
        multi_cell_pearson_list = pickle.load(f)
    ensemble_list = multi_cell_pearson_list
    with open('/data1/r20user3/shared_project/Hist2Cell/code/training/pearson_log/her2st_epoch2000_lr1e-4_GCN_onlycell_'+case+'_cell_pearson_list.pkl', 'rb') as f:
        multi_cell_pearson_list = pickle.load(f)
    gcn_list = multi_cell_pearson_list
    with open('/data1/r20user3/shared_project/Hist2Cell/code/training/pearson_log/her2st_epoch2000_lr1e-4_GCN_onlycell_'+case+'_cell_pearson_list.pkl', 'rb') as f:
        multi_cell_pearson_list = pickle.load(f)
    trans_list = multi_cell_pearson_list
    with open('/data1/r20user3/shared_project/Hist2Cell/code/training/pearson_log/her2st_epoch100_lr1e-4_densenet_onlycell_'+case+'_cell_pearson_list.pkl', 'rb') as f:
        multi_cell_pearson_list = pickle.load(f)
    densenet_list = multi_cell_pearson_list
    
    layout = go.Layout(
        title="Her2st Case "+case,  # Add figure title
        width=800,
        height=600,
        showlegend=False,  # Remove legend
        xaxis=dict(
            title="DenseNet Pearson Coefficients",  # Add x-axis title
            scaleanchor='y',  # Ensure x-axis and y-axis have the same scale
            scaleratio=1,  # Set the scale ratio to 1
        ),
        yaxis=dict(
            title="Ensemble Pearson Coefficients"  # Add y-axis title
        ),
    )

    fig = go.Figure(layout=layout)

    for x_value, y_value, label in zip(densenet_list, ensemble_list, cell_names):
        fig.add_trace(go.Scatter(x=[x_value], y=[y_value], mode='markers', marker=dict(size=10), text=[label], name=label, hovertemplate="%{text}<extra></extra>"))

    # 配置鼠标悬停信息
    fig.update_traces(hoverinfo='name')

    # 显示图像
    fig.show()
```
