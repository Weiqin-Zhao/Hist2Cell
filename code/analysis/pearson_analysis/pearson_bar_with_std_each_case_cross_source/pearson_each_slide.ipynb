```python
from glob import glob

prediction_files = glob("/data1/r20user3/shared_project/Hist2Cell/code/analysis/inference/breast_cross_source/breast_cross_source_epoch100_lr1e-4_2hop_ensemble_Trans1layer_GNNoutput50_onlycell_"+"*"+"_best_cell_all_abundance_average.pkl")
prediction_files.sort()
prediction_files
```

```python
import joblib
from scipy.stats import pearsonr
import numpy as np


slide_pearson_dict = dict()

for case_path in prediction_files:
    prediction = joblib.load(case_path)
    for slide in prediction:
        print(slide)
        
        test_cell_abundance_all_pearson_average = 0.0
        for i in range(39):
            r, p = pearsonr(prediction[slide]['cell_abundance_labels'][:, i], prediction[slide]['cell_abundance_predictions'][:, i])
            if np.isnan(r):
                r = 0.0
            test_cell_abundance_all_pearson_average = test_cell_abundance_all_pearson_average + r
        test_cell_abundance_all_pearson_average = test_cell_abundance_all_pearson_average / 39
        print(test_cell_abundance_all_pearson_average)
        
        slide_pearson_dict[slide] = test_cell_abundance_all_pearson_average
```

```python
import statistics

statistics.mean(list(slide_pearson_dict.values()))
```
