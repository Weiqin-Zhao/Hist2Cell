```python
import pickle
import statistics
```

```python
# load multi-task GNN cell pearson list

with open('/data2/r10user3/Spatial_Gene_Cell_Ratio/code/pearson_log/cell_gene_2layersimpleGNN_lr1e-4_cell_pearson_list.pkl', 'rb') as f:
    multi_cell_pearson_list = pickle.load(f)
statistics.mean(multi_cell_pearson_list)
```

```python
# get top 10 multi-task GNN cell pearson index

# Use the enumerate() function to get the index and value of each element.
indexed_list = list(enumerate(multi_cell_pearson_list))
# Use the sorted() function to sort the list and select the top ten largest values.
sorted_list = sorted(indexed_list, key=lambda x: x[1], reverse=True)[:10]
# For each of the largest values, use the index() method to get its index in the list.
max_indices = [item[0] for item in sorted_list]

max_indices
```

```python
# print the cell type name

import pandas as pd

cell_names = list(pd.read_csv('/data2/r10user3/Spatial_Gene_Cell_Ratio/data/human_lung_cell2location/WSA_LngSP8759311/cell_ratio.csv').columns)[1:]
for idx in max_indices:
    print(cell_names[idx] + ': ' + str(multi_cell_pearson_list[idx]))
```
