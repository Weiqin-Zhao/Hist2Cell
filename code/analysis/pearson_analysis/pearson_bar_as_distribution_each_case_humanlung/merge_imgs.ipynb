```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Define the directory where the images are located
image_directory = "/home/r15user3/Documents/shared_project/Hist2Cell/code/analysis/pearson_analysis/pearson_bar_as_distribution_each_case_humanlung"

# Get a list of image file names
image_files = [f for f in os.listdir(image_directory) if f.endswith('.png') and f.startswith('density')]
image_files.sort()

# Define the number of rows and columns
n_rows = (len(image_files) + 3) // 4
n_cols = 4

# Create a figure and axes
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, n_rows * 4))

# Loop through the images and plot them
for i, img_file in enumerate(image_files):
    img_path = os.path.join(image_directory, img_file)
    img = mpimg.imread(img_path)
    
    # Calculate the row and column indices
    row = i // n_cols
    col = i % n_cols
    
    # Plot the image on the corresponding axis
    axes[col].imshow(img)
    # axes[row, col].set_title(img_file)
    axes[col].axis('off')

# If there are any empty spots, remove the axes
for i in range(len(image_files), n_rows * n_cols):
    row = i // n_cols
    col = i % n_cols
    fig.delaxes(axes[row, col])

# Display the figure
plt.tight_layout()
plt.show()
```
