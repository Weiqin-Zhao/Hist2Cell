```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

ensemble_list = [] 
gcn_list = []
trans_list = []
densenet_list = []
for case in test_slides:
    with open('/data1/r20user3/shared_project/Hist2Cell/code/training/pearson_log/humanlung_epoch100_lr1e-4_2hop_ensemble_onlycell_'+case+'_cell_pearson_list.pkl', 'rb') as f:
        multi_cell_pearson_list = pickle.load(f)
    ensemble_list += multi_cell_pearson_list
    with open('/data1/r20user3/shared_project/Hist2Cell/code/training/pearson_log/humanlung_epoch2000_lr1e-4_GCN_onlycell_'+case+'_cell_pearson_list.pkl', 'rb') as f:
        multi_cell_pearson_list = pickle.load(f)
    gcn_list += multi_cell_pearson_list
    with open('/data1/r20user3/shared_project/Hist2Cell/code/training/pearson_log/humanlung_epoch2000_lr1e-4_GCN_onlycell_'+case+'_cell_pearson_list.pkl', 'rb') as f:
        multi_cell_pearson_list = pickle.load(f)
    trans_list += multi_cell_pearson_list
    with open('/data1/r20user3/shared_project/Hist2Cell/code/training/pearson_log/humanlung_epoch100_lr1e-4_densenet_onlycell_'+case+'_cell_pearson_list.pkl', 'rb') as f:
        multi_cell_pearson_list = pickle.load(f)
    densenet_list += multi_cell_pearson_list

# Sample data (replace with your own data)
data = {
    'Hist2Cell':ensemble_list, 
    'GNN':gcn_list, 
    'Transformer':trans_list, 
    'STNet':densenet_list,
}
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Create a box plot using Seaborn with custom colors
fig, ax = plt.subplots(figsize=(6, 6))
sns.boxplot(data=df, palette=colors, width=0.6, linewidth=2)

# Customize the plot
plt.title('Human Lung Case Overall', fontsize=16)
plt.xlabel("Models", fontsize=14)
plt.ylabel("Pearson R", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Remove the top and right spines
sns.despine()

plt.savefig('/data1/r20user3/shared_project/Hist2Cell/code/analysis/pearson_analysis/pearson_boxplot_with_std_each_case_humanlung/'+'overall'+'.png', dpi=300)
plt.show()
plt.close()
```
