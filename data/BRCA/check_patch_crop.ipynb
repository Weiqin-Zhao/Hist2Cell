```python
from glob import glob


patches_list = glob("/data1/r20user3/shared_project/Hist2Cell/data/BRCA/WSI/BRCA/single/BRCA/TCGA-A8-A07R-01Z-00-DX1/*")
x_coord, y_coord = [], []
for patch in patches_list:
    x_coord.append(int(patch.split('/')[-1].split('_')[0]))
    y_coord.append(int(patch.split('/')[-1].split('_')[1].split('.')[0]))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=x_coord, y=y_coord, s=5)

# 设置x和y轴的最小刻度为0
plt.xlim(xmin=0)
plt.ylim(ymin=0)

# 按照y轴翻转散点图
plt.gca().invert_yaxis()


plt.show()
```

```python
x_coord_scale, y_coord_scale = [], []
for i in range(len(x_coord)):
    x_coord_scale.append(x_coord[i]*10)
    y_coord_scale.append(y_coord[i]*10)
```

```python
import openslide
import matplotlib.pyplot as plt
from sympy import apart

# 打开SVS文件
slide = openslide.OpenSlide('/data1/r20user3/shared_project/Hist2Cell/data/BRCA/WSI/BRCA/BRCA/TCGA-A8-A07R-01Z-00-DX1.D716752E-86AF-468B-A905-A7894B978F22.svs')

# 获取缩略图
thumbnail = slide.get_thumbnail((10*max(x_coord)+70, -1))

# 使用matplotlib显示缩略图
plt.imshow(thumbnail)
# plt.axis('off')

plt.scatter(x=x_coord_scale, y=y_coord_scale, s=5, alpha=0.5)

# # 设置x和y轴的最小刻度为0
# plt.xlim(xmin=0)
# plt.ylim(ymin=0)

# 按照y轴翻转散点图
plt.gca().invert_yaxis()


plt.show()

# 关闭SVS文件
slide.close()
```
