```python
import json

with open('/data1/r20user3/shared_project/Hist2Cell/data/BRCA/rna_seq_caseid2fileid.json') as f:
    rna_json = json.load(f)
rna_json
```

```python
caseid2filename = list()
for idx in range(len(rna_json)):
    caseid2filename.append([rna_json[idx]['cases'][0]['case_id'], rna_json[idx]['file_name']])
caseid2filename[:5]
```

```python
len(caseid2filename)
```

```python
case_id_set = set()
replicate_case_id_set = set()
count = 0
for item in caseid2filename:
    if item[0] in case_id_set:
        print(item[0])
        print(item[1])
        count = count + 1
        replicate_case_id_set.add(item[0])
    else:
        case_id_set.add(item[0])
```

```python
len(replicate_case_id_set)
```

```python
caseid2filename_filtered_rep = list()
for item in caseid2filename:
    if item[0] not in replicate_case_id_set:
        caseid2filename_filtered_rep.append(item)
len(caseid2filename_filtered_rep)
```

```python
import json

with open('/data1/r20user3/shared_project/Hist2Cell/data/BRCA/wsi_caseid2fileid.json') as f:
    wsi_json = json.load(f)
wsi_json[:2]
```

```python
caseid2wsi = list()
for idx in range(len(wsi_json)):
    caseid2wsi.append([wsi_json[idx]['cases'][0]['case_id'], wsi_json[idx]['file_name']])
caseid2wsi[:5]
```

```python
case_id_set = set()
replicate_case_id_set = set()
count = 0
for item in caseid2wsi:
    if item[0] in case_id_set:
        print(item[0])
        print(item[1])
        count = count + 1
        replicate_case_id_set.add(item[0])
    else:
        case_id_set.add(item[0])
```

```python
caseid2wsi_filtered_rep = list()
for item in caseid2wsi:
    if item[0] not in replicate_case_id_set:
        caseid2wsi_filtered_rep.append(item)
len(caseid2wsi_filtered_rep)
```

```python
caseid2wsi_filtered_rep[:2]
```

```python
caseid2filename_filtered_rep[:2]
```

```python
caseid2rna_wsi = list()

for caseA in caseid2filename_filtered_rep:
    for caseB in caseid2wsi_filtered_rep:
        if caseA[0] == caseB[0]:
            caseid2rna_wsi.append([caseA[0], caseA[1], caseB[1]])
```

```python
len(caseid2rna_wsi)
```

```python
caseid2rna_wsi[:2]
```

```python
import os
from glob import glob
import shutil
from tqdm import tqdm


for case in tqdm(caseid2rna_wsi):
    wsi_id = case[2].split('.')[0]
    rna_file_name = case[1]
    os.mkdir(os.path.join('/data1/r20user3/shared_project/Hist2Cell/data/BRCA/organized_RNA_seq', wsi_id))
    des = os.path.join('/data1/r20user3/shared_project/Hist2Cell/data/BRCA/organized_RNA_seq', wsi_id, rna_file_name)
    src = glob('/data1/r20user3/shared_project/Hist2Cell/data/BRCA/RNA_seq/*/'+rna_file_name)[0]
    shutil.copy(src, des)
```

```python
all_wsi_path = glob("/data1/WSI/TCGA-WSI/BRCA/BRCA/*.svs")
for dir in tqdm(glob("/data1/r20user3/shared_project/Hist2Cell/data/BRCA/organized_RNA_seq/*")):
    for wsi_path in all_wsi_path:
        if dir.split('/')[-1] in wsi_path:
            des = os.path.join(dir, wsi_path.split('/')[-1])
            shutil.copy(wsi_path, des)
```
