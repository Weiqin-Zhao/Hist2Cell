```python
from glob import glob
import numpy as np
import pandas as pd

slide_path_list = glob("/data1/r20user3/shared_project/Hist2Cell/data/brca/*")
for slide_path in slide_path_list:
    patch_path_list = glob(slide_path + "/patches/*.jpeg")
    coord_list = list()
    for patch_path in patch_path_list:
        img_name = patch_path.split(".")[0].split("/")[-1]
        x = int(img_name.split("_")[0])
        y = int(img_name.split("_")[1])
        coord_list.append((x, y))
    spot_df = pd.DataFrame(coord_list, columns=["X", "Y"])
    spot_df.index = spot_df['X'].astype(str) + '_' + spot_df['Y'].astype(str)
    spot_df.index.rename("spot_id", inplace=True)
    spot_df.to_csv(slide_path + "/spots.csv")
```
