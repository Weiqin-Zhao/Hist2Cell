read subtyping label from the cell paper:

```python
import pandas as pd

cell_paper_df = pd.read_excel("/data1/r20user3/shared_project/Hist2Cell/data/brca_clinical_from_cell_paper/mmc2.xlsx", engine='openpyxl', skiprows=2)
cell_paper_df = cell_paper_df[['Case.ID', 'Final Pathology']]
cell_paper_df
```

read the case id list from our filtered dataset:

```python
from glob import glob

brca_filtered_case_path_list = glob("/data1/r20user3/shared_project/Hist2Cell/data/brca/*")
brca_filtered_case_id_list = [case_path.split("/")[-1] for case_path in brca_filtered_case_path_list]
brca_filtered_case_id_df = pd.DataFrame(brca_filtered_case_id_list, columns=['Slide.ID'])
brca_filtered_case_id_df['Case.ID'] = brca_filtered_case_id_df['Slide.ID'].apply(lambda x: x.split('-01Z')[0])
brca_filtered_case_id_df
```

pair the filtered case id with the subtyping label, and only keep `IDC` and `ILC` subtypes:

```python
merged_df = pd.merge(brca_filtered_case_id_df, cell_paper_df, on='Case.ID', how='left')
subtyping_df = merged_df[(merged_df['Final Pathology'] == 'IDC') | (merged_df['Final Pathology'] == 'ILC')]
subtyping_df
```

```python
subtyping_df.value_counts('Final Pathology')
```

In total, there are 465 cases left, save the paired slide id and subtype label to a csv file:

```python
subtyping_df.to_csv("/data1/r20user3/shared_project/Hist2Cell/data/brca_clinical_from_cell_paper/filtered_subtyping_label.csv", index=False)
```
